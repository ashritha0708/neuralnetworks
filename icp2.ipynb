```python
class Employee:

    empCount = 0

    def __init__(self, name, family, salary, department):
        self.name = name
        self.family = family
        self.salary = salary
        self.department = department

        Employee.empCount += 1

    def employeeCount():
        return Employee.empCount

    def average_salary(self, totalSalary, totalEmployees):
        if totalEmployees == 0:
            return 0
        return totalSalary / totalEmployees



class FulltimeEmployee(Employee):
    def __init__(self, name, family, salary, department, hoursWorked):
        super().__init__(name, family, salary, department)


        self.hoursWorked = hoursWorked


# Employee
emp1 = Employee("Ashritha", "Reddy", 50000, "Computer Science")
emp2 = Employee("Jyothsna", "Reddy", 60000, "Data Science")

# full time employee

fEmp1 = FulltimeEmployee("Abhinav", "Roy", 80000, "Artificial Intelligence", 10)
fEmp2 = FulltimeEmployee("Rohit", "Chowdary", 90000, "Cyber", 12)


print(f"Total employees: {Employee.employeeCount()}")

print(f"Employee: {emp1.name}, Family: {emp1.family}, Salary: {emp1.salary}, Department: {emp1.department}")
print(f"Employee: {emp2.name}, Family: {emp2.family}, Salary: {emp2.salary}, Department: {emp2.department}")

print(f"Fulltime Employee: {fEmp1.name}, Family: {fEmp1.family}, Salary: {fEmp1.salary}, Department: {fEmp1.department}, Hours Worked: {fEmp1.hoursWorked}")
print(f"Fulltime Employee: {fEmp2.name}, Family: {fEmp2.family}, Salary: {fEmp2.salary}, Department: {fEmp2.department}, Hours Worked: {fEmp2.hoursWorked}")

totalSalary = emp1.salary + emp2.salary + fEmp1.salary + fEmp2.salary
totalEmployees = Employee.employeeCount()
average = emp1.average_salary(totalSalary, totalEmployees)
print(f"Average salary: {average}")

```
