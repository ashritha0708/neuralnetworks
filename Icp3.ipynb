```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Read the provided CSV file ‘data.csv’
df = pd.read_csv('data.csv')

# 2. Show the basic statistical description about the data
print(df.describe())

# 3. Check if the data has null values
print(df.isnull().sum())

# 4. Replace the null values with the mean
df.fillna(df.mean(), inplace=True)

# 5. Select at least two columns and aggregate the data using: min, max, count, mean
print("\nAggregated Data for 'Calories' and 'Pulse':")
aggregated_data = df[['Calories', 'Pulse']].agg(['min', 'max', 'count', 'mean'])
print(aggregated_data)

# 6. Filter the dataframe to select the rows with calories values between 500 and 1000
filtered_df_calories = df[(df['Calories'] >= 500) & (df['Calories'] <= 1000)]
print("\nFiltered Data:")
print(filtered_df_calories)

# 7. Filter the dataframe to select the rows with calories values > 500 and pulse < 100
filteredConditions = df[(df['Calories'] > 500) & (df['Pulse'] < 100)]
print("\nFiltered Data (Calories > 500 and Pulse < 100):")
print(filteredConditions)

# 8. Create a new “df_modified” dataframe that contains all the columns from df except for “Maxpulse”
df_modified = df.drop(columns=['Maxpulse'])
print("\nNew DataFrame 'df_modified' (excluding 'Maxpulse'):")
print(df_modified.head())

# 9. Delete the “Maxpulse” column from the main df dataframe
df.drop(columns=['Maxpulse'], inplace=True)
print("\nMain DataFrame after deleting 'Maxpulse':")
print(df.head())

# 10. Convert the datatype of Calories column to int datatype
df['Calories'] = df['Calories'].astype(int)
print("\nData Types after converting 'Calories' to int:")
print(df.dtypes)

#11. Create a scatter plot for the two columns (Duration and Calories)
plt.scatter(df['Duration'], df['Calories'])
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.title('Scatter Plot of Duration vs Calories')
plt.show()


```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt


#1. Import the given salary_data.csv
data = pd.read_csv('Salary_Data.csv')
print(data.head())

#2. Split the data into training and testing partitions
FeatureData = data[['YearsExperience']]
targetData = data['Salary']

FeatureData_train, FeatureData_test, targetData_train, targetData_test = train_test_split(FeatureData, targetData, test_size=1/3, random_state=42)

#3. Train the model and make predictions

model = LinearRegression()
model.fit(FeatureData_train, targetData_train)

y_pred = model.predict(FeatureData_test)

#4. Calculate the Mean Squared Error (MSE)

mse = mean_squared_error(targetData_test, y_pred)
print(f'Mean Squared Error: {mse}')

#5. Visualize both train and test data using scatter plots:


plt.scatter(FeatureData_train, targetData_train, color='blue', label='Training data')


plt.scatter(FeatureData_test, targetData_test, color='green', label='Testing data')


plt.plot(FeatureData_test, y_pred, color='red', linewidth=2, label='Regression line')

plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()

plt.show()




```
